```python
# Is Manhattan expensive across all room types, or just entire homes?
# Result : Expensive across all room types

grouped = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()
grouped
```

```python
# Using Seaborn library to plot

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(
    data=grouped,
    x='neighbourhood_group',
    y='price',
    hue='room_type'
)
plt.title('Average Price by Neighbourhood Group and Room Type')
plt.ylabel('Average Price ($)')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.tight_layout()
#plt.ylim(0, 1000) 
plt.show()
```

```python
# Grouping by neighbourhood group and price and finding out mean, max and min prices

group_neighbour = df.groupby('neighbourhood_group')['price'].agg(['mean', 'max', 'min'])
group_neighbour
```
